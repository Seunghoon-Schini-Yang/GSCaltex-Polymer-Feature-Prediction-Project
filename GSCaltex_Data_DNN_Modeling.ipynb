```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import layers

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import math
```

```python
data = pd.read_excel('F:\\Downloads\\GS Project Big leader\\GS Data Edited_Final.xlsx')
data = data.iloc[:,:97]
```

```python
target = 'density'

data_target = data.loc[data[target].notnull()]

X = data_target.iloc[:,:90]

## X scaling options

# X = X+1
# X = np.log(X)
# X = X/math.log(101)

# X = np.sqrt(X)/10
X = X/100

y = data_target.iloc[:,90:97]

## y scaling options

y_target = y[target]
# y_target = np.sqrt(y[target])
# y_target = np.log(y[target])


from sklearn.model_selection import train_test_split
X_train, X_den_test, y_train, y_den_test = train_test_split(X, y_target, random_state=0)


def Density_model():
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[len(X.columns)]),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.Adam(0.001)
    
    model.compile(loss='mape',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
    
    return model


model_density = Density_model()

## model.summary()

run_density = model_density.fit(X_train, y_train, validation_data=[X_den_test, y_den_test] , epochs=300, batch_size=32)


y_mae = run_density.history['loss']
y_v_mae = run_density.history['val_loss']
x_len = np.arange(len(y_mae))

plt.plot(x_len, y_mae, marker='.', c='blue', label="Train-set mape")
plt.plot(x_len, y_v_mae, marker='.', c='red', label="Test-set mape")

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
## 비중 예측 실제 비교

den_pre = model_density.predict(X_den_test)

den_pred = []
for i in range(len(den_pre)):
    den_predict = den_pred.append(float(den_pre[i]))

den_pred = pd.DataFrame(den_pred)
den_pred = den_pred.set_index(y_den_test.index)

den_comp = pd.DataFrame(y_den_test)
den_comp['예측'] = den_pred
den_comp = den_comp.sort_index()

## if 스케일 되어있다면,
## 스케일 된 y 값 원래 형태로 되돌려서 보기

den_comp = np.square(den_comp)
# den_comp = np.exp(den_comp)
den_comp
```
